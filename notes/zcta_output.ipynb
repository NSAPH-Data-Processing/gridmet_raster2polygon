```python
import duckdb
conn = duckdb.connect()
```

```python
conn.execute("""
    SELECT 
        * 
    FROM 
        '../data/intermediate/sph_2020_zcta.parquet' 
    WHERE
        day = '2020-01-01'
    LIMIT 10
""").fetchdf()
```

```python
conn.execute("""
    SELECT 
        * 
    FROM 
        '../data/output/meteorology__gridmet__zcta_daily__*.parquet' 
    WHERE
        date = '2020-01-01' AND zcta = '15301'
""").fetchdf()
```

```python
conn.execute("""
    SELECT 
        * 
    FROM 
        '../data/output/meteorology__gridmet__zcta_daily__*.parquet' 
    LIMIT 10
""").fetchdf()
```

```python
conn.execute("""
    SELECT 
        count(zcta) AS nrows 
    FROM 
        '../data/output/meteorology__gridmet__zcta_daily__2020.parquet' 
""").fetchdf()
```

```python
conn.execute("""
    SELECT 
        count(distinct zcta) AS n_zcta, 
        count(distinct date) AS n_dates
    FROM 
        '../data/output/meteorology__gridmet__zcta_daily__2020.parquet' 
""").fetchdf()
```
